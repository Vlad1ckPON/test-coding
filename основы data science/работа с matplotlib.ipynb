```python
from matplotlib import pyplot as plt
```

```python
# сейчас создам простую диаграмму с годами и ввп в эти годы
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.5, 10289.7, 14958.3]
```

```python
# метод plot создает диаграмму, в функцию логично сначала передается значений по оси x, а потом y
plt.plot(years, gdp, color='green', marker='o', linestyle='solid') 
# чтобы назвать диаграмму есть функция  title
plt.title("Номинальный ВВП")
# чтобы назвать ось есть фукнция ylabel и xlabel соответственно
plt.ylabel("Млдр $")
# чтобы вывести - функция show
plt.show()
```

```python
# Первый пример столбчатого графика
```

```python
# создаю столбчатый график фильмов и количество оскаров каждого
movies = ["Энни Холл", "Бен-Гур", "Касабланка", "Ганди", "Вестсайдская история"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
# строю график с левыми x-координатами [xs] и высотами [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("Фильмы")
plt.ylabel("Количество наград")
# пометить ось x названиями фильмов в центре каждого столба
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
# посторю график распределения оценок
# мне нужно сгрупировать все оценки, но оценки от 90 и выше вместе
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5  for x in histogram.keys()], # сдвинуть столбец влево на 5
        histogram.values(), # назвачить каждому столбуц правильную высоту
        10, # и ширину в 10
        edgecolor=(0, 0, 0)) # cделать края столбца черными
# ось x будет от -5 до 105, y от 0 до 5
plt.axis([-5, 105, 0, 5]) 
# нужно сделать метки от 0 до 100
# plt.xticks(10 * i for i in range(11))
plt.xlabel("Дециль")
plt.ylabel("Число студентов")
plt.title("Распределения оценок за первый экзамен")
plt.show()
```

```python
# создаю дисперсию
varience = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1] # квадрат смещения
# суммарная ошибка
total_loss = [x + y for x, y in zip(varience, bias_squared)]
xs = [i for i, _ in enumerate(varience)]
```

```python
# метод plot можно вызвать несколько раз, чтобы показать несколько графиков на одной и той же диаграмме
# зеленая сплошная линия
plt.plot(xs, varience, "g-", label='Дисперсия')
# красная штрихпунтирная
plt.plot(xs, bias_squared, "r-.", label='Смещения^2')
# синяя пунктирная
plt.plot(xs, total_loss, 'b:', label='суммарная ошибка')
# если для каждой линии задано название, то легенда будет покзана автоматически
plt.legend()
plt.xlabel("Сложность модели")
plt.title("Компромис между смещением и дисперсией")
plt.show()
```
